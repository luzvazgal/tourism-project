```python
#Getting indexes of WEF collection
def IndexWEF(country, year):
    try:
        #Get only the information of the target country
        X=WEFdata[country]
        #For each country there is only three values for the following year [WEF update index each two year for this only use aviable info]
        X.dropna(inplace=True)
        if year==2015 or year==2014:
            i=0
            yearInfo=2015
        elif year==2017 or year==2016:
            i=1
            yearInfo=2017
        elif year==2019 or year==2018:
            i=2
            yearInfo=2019
        
        
        #Get the value info
        results=list(X[list(X.index)[0]][i][str(yearInfo)].values())
        
        #Store the info in the following variables
        cultural=results[0]
        government_prioritization=results[1]
        segurity_homocide=results[2]
        segurity_terrorism=results[3]
        natural_resources=results[4]
        price_competitiveness=results[5]
        service_infrastructure=results[6]
        visa_requirements=results[7]
        
    except:
        #if the country is not found or is diferent the year return 'na'
        cultural=np.nan
        government_prioritization=np.nan
        segurity_homocide=np.nan
        segurity_terrorism=np.nan
        natural_resources=np.nan
        price_competitiveness=np.nan
        service_infrastructure=np.nan
        visa_requirements=np.nan
        
    #return variables
    return cultural, government_prioritization, segurity_homocide, segurity_terrorism, natural_resources, price_competitiveness, service_infrastructure, visa_requirements
    
```

```python
#definir calcular la distancia
def DistancePoint(country1,country2):
    try:
        #find coordenates for country1
        objectSerch1=coordenateDF.loc[coordenateDF['name']==country1]
        latitude1=float(objectSerch1['latitude'])
        longitud1=float(objectSerch1['longitude'])
        point1=Point(longitud1,latitude1)
        
        #find coordenates for country1
        objectSerch2=coordenateDF.loc[coordenateDF['name']==country2]
        latitude2=float(objectSerch2['latitude'])
        longitud2=float(objectSerch2['longitude'])
        point2=Point(longitud2,latitude2)
        
        
        #GeoDataFrame
        points_df = gpd.GeoDataFrame({'geometry': [point1, point2]}, crs='EPSG:4326')
        points_df = points_df.to_crs('EPSG:4326')
        
        
        
    
        result=float(points_df['geometry'][0].distance(points_df['geometry'][1])*100)
    except:
        result=np.nan
    
    return result
```

```python
def wavg(country):
    Country=tourismData.loc[tourismData['Country']==country]
    
    iv=np.average(Country['Inbound Visitors'], weights=Country['Inbound Visitors'])
    distance=np.average(Country['Distance'], weights=Country['Inbound Visitors'])
    cultural=np.average(Country['Cultural'], weights=Country['Inbound Visitors'])
    gp=np.average(Country['Government_prioritization'], weights=Country['Inbound Visitors'])
    sh=np.average(Country['Segurity_homocide'], weights=Country['Inbound Visitors'])
    st=np.average(Country['Segurity_terrorism'], weights=Country['Inbound Visitors'])
    nr=np.average(Country['Natural_resources'], weights=Country['Inbound Visitors'])
    pc=np.average(Country['Price_competitiveness'], weights=Country['Inbound Visitors'])
    si=np.average(Country['Service_infrastructure'], weights=Country['Inbound Visitors'])
    vr=np.average(Country['Visa_requirements'], weights=Country['Inbound Visitors'])
    
    return iv, distance, cultural, gp, sh, st, nr, pc, si, vr
    
    
    
    
  
    
    
```

```python
def graphRegressSimple(x_values,y_values):
    
    #Get the variables to test hipotesis 
    
    x_values=x_values
    y_values=y_values
    name_x=x_values.name
    name_y=y_values.name
    
    
    #Estamimate linear regression simple
    (slope, intercept, rvalue, pvalue, stderr) =linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    if intercept<0:
        string='-'
    else:
        string='+'
    line_eq = "y = " + str(round(slope,2)) + f'*x {string}'  + str(round(intercept,2))
    r_value=rvalue
    
    #Construct graph sacatter
    
    plt.scatter(x_values,y_values, color='black', marker='o', facecolor='steelblue')
    plt.plot(x_values, regress_values, color='red')
    plt.annotate(f'{line_eq} with R={round(r_value,2)} and p-value:{round(pvalue,4)}',(x_values.min(),y_values.max()), fontsize='x-large', color="black", alpha=1) 
    plt.title(f'{name_y} vs {name_x}', fontsize='xx-large')
    plt.grid(lw=0.2)
    plt.ylabel(name_y, fontsize='large')
    plt.xlabel(name_x, fontsize='large')
    
    
    return plt.show()
    
    
```

## Model of the project

<hr>
<p>In this project, we start with the hypothesis that tourism is a function of distance, cultural offer, natural resources, government participation in tourism, the nation's security, price competitiveness, infrastructure, and visa requirements. Is for this motive, we estimate distance and used the index of the Travel & Tourism Competitiveness Report 2019 (WEF) by the World Economic Forum. The following is tranform in panel data the information to be used to determinate the funtion:</p>
<p>$Tourism $=$f$ ($distance$,$culture$,$GP$,$SH$,$ST$,$NR$,$PC$,$SI$,$VR$)</p>
<p>where</p>
<p>$distance$= Indicate how far are two contries by kilometers</p>
<p>$culture$= The index of cultural resources and business 7 indicate more offer of cultural resources while 1 indicate less offer of cultural resources</p>
<p>$GP$= The index Government prioritization of the travel & tourism industry 7 indicate top priority while 1 indicate no priority at all</p>
<p>$SH$= The index Homicide rate 7 less number of homocide per 100,000 population  while 1 less number of homocide per 100,000 population</p>
<p>$ST$= The Index of terrorism incidence 7 indicate no attack while 1 most attacks or casualities</p>
<p>$NR$= Natural Resources 7 indicate more offer of natural resources while 1 indicate less offer</p>
<p>$PC$= Price competitiveness 7 indicate more price competitiviness while 1 less price competitiviness</p>
<p>$SI$= The transport infraestructura 7 indicate more developet infraestructure while 1 less develop infraestructure</p>
<p>$VR$= Visa requirements two countries 7 less requirements while 1 more requirements</p>

```python
def ResultsModel(X_Train, X_Test, y_Train, y_Test,X_var):
    XTrain=X_Train
    XTest=X_Test
    yTrain=y_Train
    yTest=y_Test
    try:
        X=list(X_var.columns)
    except:
        X=list(X_var.name)
    #Compute linear model
    modelMultilinar=linear_model.LinearRegression().fit(XTrain,yTrain)
    yPredict=modelMultilinar.predict(XTest)
    predict=yPredict
    compare=(predict-yTest)
    R2=r2_score(yTest, yPredict)
    
    #Have the data frame with the result of model
    var=X
    coef=list(modelMultilinar.coef_)
    var.append('Intercept')
    coef.append(modelMultilinar.intercept_)
    
    
    Params=pd.DataFrame({
    'Variables':var,
    'Coeficientes':coef
    })
       
    

    return Params

    
```

```python
 def graphModelResults(X_Train, X_Test, y_Train, y_Test):
    
    XTrain=X_Train
    XTest=X_Test
    yTrain=y_Train
    yTest=y_Test
    
    #Compute linear model
    modelMultilinar=linear_model.LinearRegression().fit(XTrain,yTrain)
    yPredict=modelMultilinar.predict(XTest)
    predict=yPredict
    compare=(predict-yTest)
    R2=r2_score(yTest, yPredict)
    
    #Graph
    plt.scatter(predict,compare, color='black', marker='o', facecolor='steelblue')
    plt.title(f'Predictions vs Real Data', fontsize='xx-large')
    plt.hlines(y=0, xmin=predict.min(), xmax=predict.max(), color='red')  
    

    return print(round(R2,3)), plt.show()
    
    
    
    
    
    
```

```python
DataforPD=tourismData.set_index(['Country','Year'])
```

```python
def GraphPD(Y_Predict, Y_TestPD):
    predict=Y_Predict
    real=Y_TestPD
    predictY=predict.reset_index()
    realY=real.reset_index()


    
    x=predictY['predictions']
    Ry=realY['Inbound Visitors']
    y=(x-Ry)
    plt.scatter(x,y, color='black', marker='o', facecolor='steelblue')
    plt.title(f'Predictions vs Real Data', fontsize='xx-large')
    plt.hlines(y=0, xmin=x.min(), xmax=x.max(), color='red')
    
    
    return plt.show()
```
