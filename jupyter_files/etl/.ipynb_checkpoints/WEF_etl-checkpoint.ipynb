## Cleaning Data

#### Remove unnecessary columns

```python
deleted_columns = ['Unnamed: 0', 'Placement', 'Freeze date', 'Dataset',  'GLOBAL ID', 'Code', 
                   'Series indented',
                   'Sample average',
'Sub-Saharan Africa',
'Americas',
'Middle East and North Africa',
'Asia-Pacific',
'Europe and Eurasia ',
'North and Central America',
'South America',
'South Asia',
'Eastern Asia and Pacific',
'South-East Asia',
'Balkans and Eastern Europe',
'Eurasia',
'Northern Europe',
'Southern Europe',
'Western Europe',
'North Africa',
'Middle East',
'Western Africa',
'Eastern Africa',
'Southern Africa',
'Upper-middle-income',
'Lower-middle-income',
'High-income',
'Low-income', 'Attribute']
```

```python
wef_df.drop(labels=deleted_columns, axis=1, inplace=True)
```

```python
wef_df = wef_df.sort_values(by=['Edition', 'Series'])
```

```python
wef_df.columns = wef_df.columns.str.replace('.', '', regex=False)
```

```python
wef_df_2015 = wef_df.loc[wef_df['Edition']==2015]
wef_dict_2015 = wef_df_2015.drop(labels='Edition', axis=1).set_index(keys='Series').to_dict(orient='dict')
```

```python
wef_df_2017 = wef_df.loc[wef_df['Edition']==2017]
wef_dict_2017 = wef_df_2017.drop(labels='Edition', axis=1).set_index(keys='Series').to_dict(orient='dict')
```

```python
wef_df_2019 = wef_df.loc[wef_df['Edition']==2019]
wef_dict_2019 = wef_df_2019.drop(labels='Edition', axis=1).set_index(keys='Series').to_dict(orient='dict')
```

```python
wef_dict_final = []

for country in wef_dict_2015.keys():
    dict_2015 = {'2015': wef_dict_2015[country]}
    dict_2017 = {'2017': wef_dict_2017[country]}
    dict_2019 = {'2019': wef_dict_2019[country]}
    
   
    wef_dict_final.append({country:[dict_2015, dict_2017, dict_2019]})
    #wef_dict_final.append({country: {{'2015': wef_dict_2015[country]}, 
                            #{'2017': wef_dict_2017[country]}, 
                                     #{'2019': wef_dict_2019[country]}} })
wef_dict_final
```
