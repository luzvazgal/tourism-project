# World Economic Forum data ETL

World Economic Forum Transform (Cleaning) process and Loading into Mongo DB

```python
import pandas as pd
```

```python
wef_path = "../../data/wef_data/WEF.xlsx"
wef_df = pd.read_excel(wef_path, header=3)
```

```python
wef_df
```

## Cleaning Data

#### Remove unnecessary columns

```python
deleted_columns = ['Unnamed: 0', 'Placement', 'Freeze date', 'Dataset', 'Edition', 'GLOBAL ID', 'Code', 
                   'Series indented',
                   'Sample average',
'Sub-Saharan Africa',
'Americas',
'Middle East and North Africa',
'Asia-Pacific',
'Europe and Eurasia ',
'North and Central America',
'South America',
'South Asia',
'Eastern Asia and Pacific',
'South-East Asia',
'Balkans and Eastern Europe',
'Eurasia',
'Northern Europe',
'Southern Europe',
'Western Europe',
'North Africa',
'Middle East',
'Western Africa',
'Eastern Africa',
'Southern Africa',
'Upper-middle-income',
'Lower-middle-income',
'High-income',
'Low-income']
```

```python
wef_df.drop(labels=deleted_columns, axis=1, inplace=True)
wef_df
```

Retrieving rows with Attribute == 'Value'

```python
wef_df_short = wef_df.loc[wef_df['Attribute']=='Value']
```

```python
wef_df_short.drop_duplicates(subset='Series',keep='first', inplace=True)
wef_df_short
```

Removing '.' from column names

```python
wef_df_short.columns = wef_df_short.columns.str.replace('.', '', regex=False)
```

Dropping columns: 'Attribute', 'Series'

```python
index_names = wef_df_short['Series'].str.replace('.', '', regex=False)
```

```python
for name in index_names:
    print(name)
```

```python
wef_df_short.drop(labels=['Attribute', 'Series'], axis=1, inplace=True)
```

Assigning Series values as indices

```python
wef_df_short.index = index_names
```

```python
wef_df_short
```

Converting data frame to dictionary to be ingested into Mongo DB as a collection

```python
wef_dict = wef_df_short.to_dict(orient='dict')
```

```python
wef_dict
```

```python
print(type(wef_dict))
```

```python
wef_list = []

for key, value in wef_dict.items(): 
    wef_list.append({key:value})
    
wef_list
```
